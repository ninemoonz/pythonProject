```python
import pandas as pd
import pymysql

# csv file
sts = pd.read_csv('/Users/followthesnake/Desktop/data/state.csv')
print(sts.info())

state = sts['State']
pop = sts['Population']
mur = sts['Murder.Rate']
abbr = sts['Abbreviation']

config = {
    'host' : 'localhost',
    'user' : 'root',
    'password' : '',
    'database' : 'pythondb',
    'port' : 3306,
    'charset' : 'utf8',
    'use_unicode' : True
}

try:
    conn = pymysql.connect(**config)
    cursor = conn.cursor()

    # table 조회
    cursor.execute("show tables")
    tables = cursor.fetchall()

    # switching 기법
    sw = False
    for table in tables:
        if table[0] == 'states_tab':
            sw = True # table 있는 경우 swapping

    if not sw :
        print('no table')
        sql="""create table states_tab(
        states varchar(30),
        population int not null,
        murder_rate int not null,
        abbreviation varchar(5)
        )"""
        cursor.execute(sql)

    # value search
    cursor.execute("select * from states_tab")
    rows = cursor.fetchall()

    if rows:
        for row in rows :
            print(f"{row[0]}    {row[1]}    {row[2]}    {row[3]}")
        print('Number of whole record: ', len(rows))
    else:
        print("Adding 100 values")
        for i in range(100):
            s = state[i]
            p = pop[i]
            m = mur[i]
            a = abbr[i]
            cursor.execute(f"insert into states_tab (state, pop, mur, abbr) values ('{s}', {p}, {m}, '{a}')")
            conn.commit()

except Exception as e:
    print('db error: ', e)
finally:
    cursor.close()
    conn.close()
```
