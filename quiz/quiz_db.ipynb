```python
import pymysql

config = {
    'host': 'localhost',
    'user': 'root',
    'password': '',
    'database': 'pythondb',
    'port': 3306,
    'charset': 'utf8',
    'use_unicode': True
}

try:
    conn = pymysql.connect(**config)
    cursor = conn.cursor()

    #Loop
    while True:
        print('\t[Record Processing Menu]')
        print('1. Record Search')
        print('2. Record Insert')
        print('3. Record Edit')
        print('4. Record Delete')
        print('5. Program Exit')

        menu = int(input('\tEnter Menu Number: '))

        if menu == 1:
            print('>> Going over all the goods <<')
            sql = "select * from goods"
            cursor.execute(sql)
            tables = cursor.fetchall()

            for row in tables:
                print(row[0], row[1], row[2], row[3])

        elif menu == 2: # insert

            code = int(input('enter code number: '))
            name = input('enter the name of appliance: ')
            stock = int(input('enter the number of stock: '))
            price = int(input('enter the price of appliance: '))

            sql = "insert into goods (code, name, stock, price) values (%d, '%s', %d, %d)" %(code, name, stock, price)
            cursor.execute(sql)
            conn.commit()

        elif menu == 3: # edit
            code = int(input('editing code: '))
            name = input('Name to change: ')
            size = int(input('change in stock: '))
            portion = int(input('change in price: '))

            sql = f"update goods set name = '{name}', stock = {stock}, price = {price} where code = {code}"
            cursor.execute(sql)
            conn.commit()

        elif menu == 4: # delete
            code = int(input('Code of deleting product: '))
            sql = f"select * from meds where code = {code}"
            rows = cursor.execute(sql)

            if rows :
                print('delete record')
                sql = f"delete from goods where code = {code}"
                cursor.execute(sql)
                conn.commit()

        elif menu == 5:
            print('Exiting the Program')
            break
        else:
            print('No such order exist')

except Exception as e:
    print('db error: ', e)
    conn.rollback()

finally:
    conn.close()
    cursor.close()
```

```python
import pymysql
import pandas as pd

config = {
    'host': 'localhost',
    'user': 'root',
    'password': '',
    'database': 'pythondb',
    'port': 3306,
    'charset': 'utf8',
    'use_unicode': True
}

try:
    conn = pymysql.connect(**config)
    cursor = conn.cursor()

    # condition 1: searching through height between 170 ~ 180
    sql = "select * from bmi_tab where height >= 170 and height <= 180"
    cursor.execute(sql)
    rows = cursor.fetchall()

    for row in rows:
        print(row[0], row[1], row[2])

    # condition 2: searching with label
    label = input('Searching label: ')
    sql = f"select * from bmi_tab where label like \'%{label}%\'"
    cursor.execute(sql)
    labels = cursor.fetchall()

    if rows:
        for label in labels:
            print(label)
        print('current label record number: ', len(rows))
    else:
        print('no such label exists')


except Exception as e:
    print('db error: ', e)
    conn.rollback()
finally:
    conn.close()
    cursor.close()
```
